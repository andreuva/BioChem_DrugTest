```python
#https://www.youtube.com/watch?v=arYKpiVACvY&ab_channel=SignalProcessingwithPaul
import numpy as np, matplotlib.pyplot as plt
```

```python
#define x(t) as as pulse function
num_t = 100
xt = np.zeros(num_t)
xt[:10] = 1
#xt[:5] = 1
#xt[-4:] = -1
plt.plot(xt)
plt.title("PLot of x(t)")
plt.xlabel("Sample index")
plt.ylabel("Value of x(t)")
```

```python
fxt = np.fft.fft(xt)
#First plot is the real plot
plt.subplot(1,2,1)
plt.plot(np.fft.fftshift(np.real(fxt)))
plt.title("Real part x_hat(t) FFT")
plt.ylabel("Magnitud")
plt.xlabel("Frequency Index")
# imaginarty plat
plt.subplot(1,2,2)
plt.plot(np.fft.fftshift(np.imag(fxt)))
plt.title("Img. part x_hat(t) FFT")
plt.ylabel("Magnitud")
plt.xlabel("Frequency Index")
```

```python
# Plot in polar form
fxt = np.fft.fft(xt)
#First plot is the real plot
plt.subplot(1,2,1)
plt.plot(np.fft.fftshift(np.abs(fxt)))
plt.title("Real part x_hat(t) FFT")
plt.ylabel("Magnitud")
plt.xlabel("Frequency Index")
# imaginarty plat
plt.subplot(1,2,2)
plt.plot(np.fft.fftshift(np.angle(fxt)))
plt.title("Img. part x_hat(t) FFT")
plt.ylabel("Angle")
plt.xlabel("Frequency Index")
```
