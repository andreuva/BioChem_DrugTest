# Modelos de ML

```python
import pandas as pd
```

```python
path = 'data/all_metrics.csv'
df_all = pd.read_csv(path,index_col=0)
df_all
```

## Modelo basico

Usando
1) Media
2) Mediana
3) Desv. Estandar
4) Q1 y Q3

```python
columns_for_model = ['Cycle', 'Area [%]', 'Perimeter [um]', 'Form factor', 'Extent', 'Compactness','Eccentricity', 
                     'Characteristic RI','Characteristic dry mass density [pg / um^3]', 'Dry mass [pg]', 'Granularity']
columns_to_predict = ['Target']
```

```python
df_all['ID'] = df_all['Well'] + '_' + df_all['Cycle'].astype(str) + '_' + df_all['Metric'].astype(str)
df_all.head()
```

```python
len(set(df_all['ID']))
```

```python
df_group = df_all.groupby('ID').min().reset_index()
df_group[columns_for_model+columns_to_predict]
```

```python
# Example column names - replace 'column1', 'column2' with your actual column names
columns_for_aggregation = columns_for_model

# Grouping by 'ID' and applying multiple aggregation functions to the specified columns
new_df = df_all.groupby('ID').agg({
    col: ['min', 'mean', 'max', 'std', 'median', lambda x: x.quantile(0.25), lambda x: x.quantile(0.75)] for col in columns_for_aggregation
}).reset_index()

# Flattening the MultiIndex in columns
new_df.columns = ['_'.join(col).strip() if col[1] else col[0] for col in new_df.columns.values]

# Accessing the DataFrame; adjust column names based on the result of the aggregation
new_df
```

```python
new_df['Target'] = df_group['Target']
```

```python
columns_metrics_for_model = [y+'_'+x for y in columns_for_model for x in ['min', 'mean', 'max', 'std', 'median', '<lambda_0>', '<lambda_1>']]
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report


X = new_df[columns_metrics_for_model]  # Features
y = new_df[columns_to_predict]  # Target variable

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Estandarización de características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Entrenamiento del modelo
sgd_classifier = SGDClassifier(max_iter=1000, tol=1e-3,random_state=42)
sgd_classifier.fit(X_train_scaled, y_train)
```

```python
# Evaluación del modelo
y_pred = sgd_classifier.predict(X_test_scaled)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```
